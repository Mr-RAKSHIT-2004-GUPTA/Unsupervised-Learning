```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
#
```

```python
from sklearn import datasets
data = datasets.load_iris()
X = data.data
y=data.target
```

```python
X
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_scaled = sc.fit_transform(X)
```

```python
X_scaled
```

```python
#applying the pca
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca_scaled = pca.fit_transform(X_scaled)
```

```python
plt.scatter(pca_scaled[:,0] , pca_scaled[:,1],c = y)
```

```python
#agglomerative clustring >> make dedogram

import scipy.cluster.hierarchy as sch

plt.figure(figsize = (5,6))
plt.figure("dendogram")
sch.dendrogram(sch.linkage(pca_scaled,method = 'ward'))
plt.xlabel("sample data")
plt.ylabel("Ecludian distance")
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=3,metric='euclidean',linkage='ward')
cluster.fit_predict(pca_scaled)
```

```python
cluster.labels_
```

```python
plt.scatter(pca_scaled[:,0] , pca_scaled[:,1],c = cluster.labels_)
```

```python
#k starts from 2
from sklearn.metrics import silhouette_score
silhouette_coeff = []
for k in range(2,11):
  agglo = AgglomerativeClustering(n_clusters=k,metric='euclidean',linkage='ward')
  labels = agglo.fit_predict(pca_scaled)
  silhouette_coeff.append(silhouette_score(pca_scaled,labels))
```

```python
plt.plot(range(2,11),silhouette_coeff)
plt.xlabel("number of clusters")
plt.ylabel("silhouette coefficient")
```
