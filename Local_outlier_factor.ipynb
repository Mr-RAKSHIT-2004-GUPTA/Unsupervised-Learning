```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.neighbors import LocalOutlierFactor
from sklearn.datasets import make_moons

X,y = make_moons(n_samples = 300 , noise =0.1)
```

```python
X_outlier = np.random.uniform(low = 3 , high=4 , size = (20,2))
X = np.vstack((X,X_outlier))
```

```python
out = LocalOutlierFactor(n_neighbors = 20,contamination=0.1)
y_pred = out.fit_predict(X)
```

```python
y_pred
```

```python
X_inlier = X[y_pred == 1]
X_outlier = X[y_pred == -1]

#plotting
plt.title("local outlier factor(LOF) on moons data")
plt.scatter(X_inlier[:,0],X_inlier[:,1],color = 'blue',label = 'inlier',edgecolors='k')
plt.scatter(X_outlier[:,0],X_outlier[:,1],color = 'red',label = 'outlier',edgecolors='k')
plt.legend()
plt.show()
```
